# Json vs CSV

### Json

```python
url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=BTC&apikey=" + api_key
response = requests.get(url)
# print(response.status_code) #returns a code number
print(response.ok) #returns a boolean (True / False)
data = response.content
data
```

### CSV

```python
url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=BTC&apikey=" + api_key
response = requests.get(url)
if response.ok:
    with open("btc_data.csv", "wb") as f:
        f.write(response.content)

print("CSV Saved to btc_data.csv")
```
