# Web Scrapping

```python
import requests
import bs4

def scrape_data():
    url = "https://u.today/search/node?keys=Bitcoin"
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    print(soup)
    return soup

data = scrape_data()
```

```python
search_results = data.find('div', class_='search-result')
articles = search_results.find_all('div', class_='news__item')
dates_scraped = []
titles_scraped = []
authors_scraped = []
hrefs_scraped = []

for article in articles:
    date_tag = article.find('div', class_='humble')
    date_scraped = date_tag.get_text(strip=True) if date_tag else None
    dates_scraped.append(date_scraped)
    title_tag = article.find('a', class_='news__item-body')
    title_scraped = title_tag.get_text(strip=True) if title_tag else None
    titles_scraped.append(title_scraped)
    author_tag = article.find('a', class_='humble humble--author')
    author_scraped = author_tag.get_text(strip=True) if author_tag else None
    authors_scraped.append(author_scraped)
# print(dates_scraped)
# print(dates_scraped)
# print(dates_scraped)

df_scrape = pd.DataFrame({
    'date': dates_scraped,
    'title': titles_scraped,
    'author': authors_scraped,
})

df_scrape['date'] = pd.to_datetime(df_scrape['date'])
df_scrape = df_scrape.sort_values('date', ascending=False).reset_index(drop=True)

print(df_scrape.head())

df_scrape.to_csv("/Users/kimberleybrown/Desktop/coding_all/Code_Academy_Berlin/Module3/data/sentiment_data.csv", index=False)
```
